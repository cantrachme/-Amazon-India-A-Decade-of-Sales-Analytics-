```python
# 3)


import numpy as np
import re #Regular expression 


# Isolate the column

rating_col = df['customer_rating'].astype(str).copy()


# Clean & standardize


def parse_rating(value):
    value = value.lower().strip()
    
    # Handle missing-like values
    if value in ['nan', 'none', '', 'null']:
        return np.nan
    
    # Case 1: simple numeric (e.g., "5.0", "4")
    if re.fullmatch(r'\d+(\.\d+)?', value):
        return float(value)
    
    # Case 2: "4 stars"
    star_match = re.search(r'(\d+(\.\d+)?)\s*star', value)
    if star_match:
        return float(star_match.group(1))
    
    # Case 3: fraction formats "3/5", "2.5/5.0"
    frac_match = re.search(r'(\d+(\.\d+)?)/(\d+(\.\d+)?)', value)
    if frac_match:
        numerator = float(frac_match.group(1))
        denominator = float(frac_match.group(3))
        if denominator != 0:
            return (numerator / denominator) * 5
    
    return np.nan

# Apply parsing
rating_clean = rating_col.apply(parse_rating)

# Enforce valid scale 1.0–5.0
rating_clean = rating_clean.where(
    (rating_clean >= 1.0) & (rating_clean <= 5.0),
    np.nan
)


# Update original DataFrame
df['customer_rating'] = rating_clean


# VERIFICATION OUTPUT

print("Total rows:", len(df))
print("Valid ratings:", df['customer_rating'].notna().sum())
print("Missing / invalid ratings:", df['customer_rating'].isna().sum())

print("\nSample cleaned ratings:")
print(df['customer_rating'].dropna().sample(5, random_state=42))

print("\nRating distribution:")
print(df['customer_rating'].value_counts(bins=5).sort_index())

print("\nData type after cleaning:")
print(df['customer_rating'].dtype)
```

```python
# 7)



import numpy as np
import re


# Isolate the column

delivery_col = df['delivery_days'].astype(str).copy()


# Parse & clean values

def parse_delivery_days(value):
    value = value.lower().strip()
    
    # Missing-like values
    if value in ['nan', 'none', '', 'null']:
        return np.nan
    
    # Same-day delivery
    if 'same' in value:
        return 0
    
    # Range like "1-2 days"
    range_match = re.search(r'(\d+)\s*-\s*(\d+)', value)
    if range_match:
        return int(range_match.group(1))
    
    # Pure numeric (including negative)
    if re.fullmatch(r'-?\d+', value):
        days = int(value)
        if days < 0:
            return np.nan
        return days
    
    return np.nan

delivery_clean = delivery_col.apply(parse_delivery_days)


# Remove unrealistic values

delivery_clean = delivery_clean.where(
    (delivery_clean >= 0) & (delivery_clean <= 30),
    np.nan
)


# Update original DataFrame

df['delivery_days'] = delivery_clean


# VERIFICATION OUTPUT 

total_rows = len(df)
valid_count = df['delivery_days'].notna().sum()
invalid_count = df['delivery_days'].isna().sum()

print("Total rows:", total_rows)
print("Valid delivery days:", valid_count)
print("Invalid / unrealistic entries:", invalid_count)

print("\nSample cleaned delivery days:")
if valid_count > 0:
    print(
        df['delivery_days']
        .dropna()
        .sample(min(5, valid_count), random_state=42)
    )
else:
    print("No valid delivery_days values after cleaning.")

print("\nDelivery days distribution:")
print(df['delivery_days'].value_counts().sort_index())

print("\nData type after cleaning:")
print(df['delivery_days'].dtype)
```

```python
# 9)



import numpy as np
import pandas as pd


# Isolate price column

price_col = df['original_price_inr'].copy()


# IQR-based outlier detection

Q1 = price_col.quantile(0.25)
Q3 = price_col.quantile(0.75)
IQR = Q3 - Q1

upper_bound = Q3 + 1.5 * IQR

outlier_mask = price_col > upper_bound

print("Detected price outliers:", outlier_mask.sum())


# Domain-based correction

price_corrected = price_col.copy()

# IMPORTANT FIX: use pandas Series (index-aligned)
correction_flag = pd.Series(0, index=df.index)

for idx in price_col[outlier_mask].index:
    value = price_col.loc[idx]
    
    # Try ÷100 correction
    if value % 100 == 0 and (value / 100) <= upper_bound:
        price_corrected.loc[idx] = value / 100
        correction_flag.loc[idx] = 1
    
    # Try ÷10 correction
    elif value % 10 == 0 and (value / 10) <= upper_bound:
        price_corrected.loc[idx] = value / 10
        correction_flag.loc[idx] = 1
    
    # Unrecoverable outlier
    else:
        price_corrected.loc[idx] = np.nan
        correction_flag.loc[idx] = 2


# Update main DataFrame

df['original_price_inr'] = price_corrected
df['price_outlier_flag'] = correction_flag


# VERIFICATION OUTPUTS

print("\nPrice outlier handling summary:")
print(df['price_outlier_flag'].value_counts().rename({
    0: "Normal",
    1: "Corrected",
    2: "Removed"
}))

print("\nSample corrected prices:")
corrected = df[df['price_outlier_flag'] == 1]['original_price_inr']
if not corrected.empty:
    print(corrected.sample(min(5, len(corrected)), random_state=42))
else:
    print("No corrected prices found.")

print("\nSample removed prices:")
removed = df[df['price_outlier_flag'] == 2]['original_price_inr']
if not removed.empty:
    print(removed.head())
else:
    print("No prices removed.")
```

```python
import pandas as pd
import numpy as np


# 1. Ensure order_date is datetime

df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')


# 2. Create RFM base table

analysis_date = df['order_date'].max() + pd.Timedelta(days=1)

rfm = (
    df
    .groupby('customer_id')
    .agg(
        Recency=('order_date', lambda x: (analysis_date - x.max()).days),
        Frequency=('transaction_id', 'nunique'),
        Monetary=('final_amount_inr', 'sum')
    )
    .reset_index()
)


# 3. RFM SCORING 


# Rank-based scoring avoids duplicate bin issues
rfm['R_score'] = (
    rfm['Recency']
    .rank(method='first', ascending=False)
    .pipe(lambda x: pd.qcut(x, 5, labels=[5,4,3,2,1]))
)

rfm['F_score'] = (
    rfm['Frequency']
    .rank(method='first', ascending=True)
    .pipe(lambda x: pd.qcut(x, 5, labels=[1,2,3,4,5]))
)

rfm['M_score'] = (
    rfm['Monetary']
    .rank(method='first', ascending=True)
    .pipe(lambda x: pd.qcut(x, 5, labels=[1,2,3,4,5]))
)


# 4. FINAL CLEANING (CRITICAL STEP)


rfm[['R_score', 'F_score', 'M_score']] = (
    rfm[['R_score', 'F_score', 'M_score']]
    .astype(float)
    .fillna(3)          # neutral score
    .astype(int)
)


# 5. Combined RFM Score

rfm['RFM_score'] = (
    rfm['R_score'] * 100 +
    rfm['F_score'] * 10 +
    rfm['M_score']
)


# 6. Output check

print("RFM table shape:", rfm.shape)
print("\nSample RFM rows:")
print(rfm.head())
```
